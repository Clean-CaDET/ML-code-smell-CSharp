```python
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel('DataSet_Long Method - Round 3.xlsx')
```

```python
cleanup_nums = {"Final annotation":     {1: 1, 2: 1, 3: 1, 4: 1, 5: 1, 0: 0}}

df= df.replace(cleanup_nums)
```

```python
df.set_index('Code Snippet ID')
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test, y_train, y_test = train_test_split(df, df['Final annotation'],stratify=df['Final annotation'], test_size=0.20, random_state=0)
```

```python
with open('Long_Method_train_set.pickle', 'wb') as output:
    pickle.dump(df_train, output)
    
with open('Long_Method_train_set_y.pickle', 'wb') as output:
    pickle.dump(y_train, output)
    
with open('Long_Method_test_set.pickle', 'wb') as output:
    pickle.dump(df_test, output)

with open('Long_Method_test_set_y.pickle', 'wb') as output:
    pickle.dump(y_test, output)
```
